```python
import pandas as pd
import numpy as np
import nltk
import matplotlib.pyplot as plt
import re
import seaborn as sns
```

```python
data=pd.read_csv('data.csv')
data=data[~data['cleaned_text'].isnull()]
data.drop(columns=['Unnamed: 0'],inplace=True)
data.head()
```

```python
sns.set(rc={'figure.figsize':(15,15)},font_scale=2.5)
ax=sns.countplot(x="topic", data=data)
ax.set_xticklabels(ax.get_xticklabels(),rotation=90,fontsize = 18)
plt.show()
```

```python
#word counting for each text
def CountWord(text):
    return len(text.split())

data['word count']=data['cleaned_text'].apply(CountWord)
data.head()
```

```python
data.describe()
```

```python
sns.set(rc={'figure.figsize':(6,6)},font_scale=2.5)
sns.boxplot(x='word count',data=data[data['word count']<1000],orient='v')
```
